### 3. Analyze the Data

I will use visualizations to better analyse and compare the data

```python
import matplotlib.pyplot as plt
```

```python
first12_1948 = unrate[0:12]
plt.plot(first12_1948['DATE'], first12_1948['VALUE'])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

From my exploration of the unemployment data of 1948 I can see the following:

- The unemployment rate fluctuated throughout the year, starting low at about 3.4 and ending at about 4.0.
- There are two major swings in the graph that may be attributed to seasonal trends. One way to check for this is to see if these trends are shared by the unemplyment trends of other years. 

```python
fig = plt.figure(figsize=(12,12))

plt.title("Monthly Unemployment Trends, 1948-1952")

for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['VALUE'])

plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month

fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)

plt.title("Monthly Unemployment Trends, 1948-1952")   
plt.legend(loc='upper left')

plt.show()
```

From my exploration of the unemployment data of 1948-1952 I can see the following:

- The unemployment rate fluctuated throughout the years, between about 3.2 and reaching just under 8.0 at its peak in October of 1949.
- There is no clear visual representation in the data to suggest clear seasonal unemploymeny swings other than perhaps the decrease in unemployment in the warmer months, and in the holidays for some of the years represented in this graph.
- There are two major swings in the graph but for the most part, during the years of 1948-1952 the unemployment rate stayed below the normal rate of 5%.

```python
unrate['MONTH'] = unrate['DATE'].dt.month
fig = plt.figure(figsize=(10,6))

plt.plot(unrate[24:36]['MONTH'], unrate[24:36]['VALUE'], c='red')
plt.plot(unrate[36:48]['MONTH'], unrate[36:48]['VALUE'], c='blue')
plt.plot(unrate[48:60]['MONTH'], unrate[48:60]['VALUE'], c='green')
plt.plot(unrate[60:72]['MONTH'], unrate[60:72]['VALUE'], c='orange')
plt.plot(unrate[72:84]['MONTH'], unrate[72:84]['VALUE'], c='black')

plt.title("Monthly Unemployment Trends, 1950-1954")
plt.legend(['1950','1951', '1952', '1953', '1954'], loc='upper left')

plt.show()
```

From my exploration of the unemployment data of 1950-1954 I can see the following:

- Similarly from above, the unemployment rate fluctuated throughout the years, between about 2.5 and reaching just around 6.5 at its peak in January of 1950.
- Similarly from above, there appears to be a decrease in unemployment in the warmer months, and in the holidays for some of the years represented in this graph.
- There are two outliers in 1950 and 1954, but for the most part, during the years of 1950-1954 the unemployment rate stayed well below the normal rate of 5%.
