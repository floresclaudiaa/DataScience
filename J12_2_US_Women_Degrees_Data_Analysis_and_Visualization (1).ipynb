### 1. Explore the Data

```python
import pandas as pd
import matplotlib.pyplot as plt

womenb = pd.read_csv('women_bac_1970_2012.csv')

print(womenb.info())
```

### Analyze the Data

I will begin my analysis by exploring stem field data in the following columns:

- Biology
- Computer Science
- Engineering
- Math and Statistics
- Physical Sciences

```python
womenb['Biology'].describe()
```

- In the first period, from 1970 to around 1987, women were a minority when it came to majoring in Biology.
- In the second period, from around 1987 to around 2012, women became a majority. 
- You can see the point where women overtook men where the lines intersect.

```python
fig, ax = plt.subplots()
ax.plot(womenb['Year'], womenb['Biology'], c="green", label='Women')
ax.plot(womenb['Year'], 100 - womenb['Biology'], c="orange", label='Men')

ax.tick_params(bottom=False, top=False, left=False, right=False)
for key,spine in ax.spines.items():
    spine.set_visible(False)

ax.legend(loc='upper right')
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
plt.show()
```

From my brief analysis of the STEM degree columns, I can see the following:

- The data for 4 out of the 5 STEM degrees shows a significant gap between women and men representation in those fields.

As a result, I will continue my analysis by examining gender representation in the following 4 groups:

- STEM fields
- Fields represented heavily by women
- Fields that are not under the STEM umbrella, but tend to reap higher wages than liberal arts and humanities majors.

```python
stem = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']

fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(womenb['Year'], womenb[stem[sp]], c='green', label='Women')
    ax.plot(womenb['Year'], 100-womenb[stem[sp]], c='orange', label='Men')
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(stem[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.legend(loc='upper right')
plt.show()
```

```python
other = ['Education', 'Health Professions', 'Psychology', 'Public Administration']

fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(womenb['Year'], womenb[other[sp]], c='green', label='Women')
    ax.plot(womenb['Year'], 100-womenb[other[sp]], c='orange', label='Men')
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(other[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.legend(loc='upper right')
plt.show()
```

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(womenb['Year'], womenb[stem_cats[sp]], c='green', label='Women', linewidth=3)
    ax.plot(womenb['Year'], 100-womenb[stem_cats[sp]], c='orange', label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])    
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    if sp == 0:
        ax.text(2005, 87, "Men")
        ax.text(2002, 8, "Women")
    elif sp == 5:
        ax.text(2005, 62, "Men")
        ax.text(2001, 35, "Women")
plt.show()
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']


fig = plt.figure(figsize=(16, 16))

## stem
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(womenb['Year'], womenb[stem_cats[cat_index]], c=(44/255, 160/255, 44/255), label='Women', linewidth=3)
    ax.plot(womenb['Year'], 100-womenb[stem_cats[cat_index]], c=(255/255, 128/255, 14/255), label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(stem_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if cat_index == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif cat_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
        ax.tick_params(labelbottom=True)
        
## lib_arts
for sp in range(1,16,3):
    cat_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(womenb['Year'], womenb[lib_arts[cat_index]], c=(44/255, 160/255, 44/255), label='Women', linewidth=3)
    ax.plot(womenb['Year'], 100-womenb[lib_arts[cat_index]], c=(255/255, 128/255, 14/255), label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(lib_arts[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if cat_index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    elif cat_index == 4:
        ax.tick_params(labelbottom=True)

## other_cats
for sp in range(2,20,3):
    cat_index = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(womenb['Year'], womenb[other_cats[cat_index]], c=(44/255, 160/255, 44/255), label='Women', linewidth=3)
    ax.plot(womenb['Year'], 100-womenb[other_cats[cat_index]], c=(255/255, 128/255, 14/255), label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if cat_index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif cat_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        ax.tick_params(labelbottom=True)

fig.savefig("gender_degrees.png")

plt.show()
```

### Conclusion

I can conclude that the gender gap in Computer Science and Engineering majors are the largest in the majors that I have examined. In addition, from the STEM degrees, Computer Science and Engineering categories are represented by men disproportionately, while Biology, and Math and Statistics degree categories are much more balanced in their gender representation.

#### How are women represented in the STEM fields?

- The data for 4 out of the 5 STEM degrees shows a significant gap between women and men representation in those fields. As a result, Biology, Computer Science, Engineering, and Math and Statistics majors are underrepresented by women.
- Physical Sciences is the only STEM field that represents both women and men more evenly after 2000 with peak representation for women at around 42% just after the year 2000. It is important to note that the representation appears to be decreasing after its peak, and asks to be further examined for the years that follow.
